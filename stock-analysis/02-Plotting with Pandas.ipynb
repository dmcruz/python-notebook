# Pandas Plot

```python
import yfinance as yf
import pandas as pd

stocklist = ['AMZN', 'GOOG', 'META', 'MSFT', 'NFLX', 'TSLA']

df = pd.DataFrame()

for ticker in stocklist:
    df[ticker] = yf.download(ticker, start='2013-09-01', end='2023-09-30', progress=False)['Close']
df.dropna(axis=0, inplace=True)
df

```

```python
df.plot(title='Stock Price over 10 Years')
```

## Subplot default format

```python
df.plot(subplots=True, grid=True)
```

## Subplots in 2x3 layout

```python
df.plot(subplots=True, layout=(2,3), title='Stock Prices over 10 Years', grid=True)
```

## Price Changes

```python
price_changes_daily = df.pct_change()

price_changes_daily.plot(subplots=True, layout=(2, 3), title="Stock Prices over 10 Years", grid=True)
```

```python
(1 + price_changes_daily).cumprod() #Compounded price change
```

```python
(1 + price_changes_daily).cumprod().plot(title='Performance of Stocks over Time', ylabel='Price', grid=True)
```
