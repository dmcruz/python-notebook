# Matplotlib Plot

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

stocklist = ['AMZN', 'GOOG', 'META', 'MSFT', 'NFLX', 'TSLA']

df = pd.DataFrame()

for ticker in stocklist:
    df[ticker] = yf.download(ticker, start='2013-09-01', end='2023-09-30', progress=False)['Close']
df.dropna(axis=0, inplace=True)

fig, ax = plt.subplots()
ax.set(title="Stock Prices over 10 Years", xlabel="Date", ylabel="Price")

for tick in stocklist:
    plt.plot(df[tick])
    
plt.show()
```

## Bar Chart

```python

price_change_daily = df.pct_change()

fig, ax = plt.subplots()
ax.set(title='Daily Return of Stocks', xlabel='Stocks', ylabel='Mean Return')
plt.bar(x=price_change_daily.columns, height=price_change_daily.mean(), color='green')

```

## Apply Theme

```python

fig, ax = plt.subplots()
plt.style.use('bmh')
ax.set(title='Daily Return of Stocks', xlabel='Stocks', ylabel='Standard Deviation')
plt.bar(x=price_change_daily.columns, height=price_change_daily.std(), color='orange')
```
